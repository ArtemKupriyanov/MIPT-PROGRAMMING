## Сжатие картинок с помощью PCA

**Задание:** построейте SVD декомпозицию <a href='http://wallpaperswide.com/under_a_yellow_tree-wallpapers.html'> картинки</a> и сожмите её в 2,4,16 и 128 раз (идею см. здесь: http://timbaumann.info/svd-image-compression-demo/).
Отобразите полученные картинки.

```python
img = cv2.imread("under_a_yellow_tree-wallpaper-1440x900.jpg")
h, w, _ = img.shape
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
k_components = int(h*w / (2 * (h + w + 1)))
k_components
```

```python
for compressed_ratio in [2, 4, 16, 128]:
    k_components = int(h*w / (compressed_ratio * (h + w + 1)))
    compressed_image = np.zeros_like(img)
    for i in range(3):
        U, s, V = np.linalg.svd(img[:, :, i], full_matrices=False)
        S = np.diag(s)

        k_components = int(h*w / (compressed_ratio * (h + w + 1)))
        svd_first_k = U[:, :k_components] @ S[:k_components, :k_components] @ V[:k_components, :]
        compressed_image[:, :, i] = svd_first_k
    
    compressed_image[compressed_image < 0] = 0
    plt.imshow(cv2.cvtColor(compressed_image, cv2.COLOR_BGR2RGB))
    plt.show()
    
    uncompressed_size = np.prod(img.shape)
    compressed_size = 3 * (np.product(U[:, :k_components].shape) + np.product(V[:k_components, :].shape) + k_components)
    
    print("compressed_ratio = {:.2f}, SVD components: {}".format(uncompressed_size / compressed_size, k_components))
```
