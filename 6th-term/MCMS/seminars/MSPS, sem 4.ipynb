```python
h0_distribution = lambda x: np.random.normal(loc = 0, size = x)
h1_distribution = lambda x: np.random.normal(loc = 2, size = x)
```

```python
def get_train_test_samples(h0_distribution=h0_distribution,
                           h1_distribution=h1_distribution,
                           sample_size = 1000,
                           h0_samples = 500,
                           h1_samples = 500):
    train = h0_distribution(sample_size)
    test = []
    labels = []
    for _ in range(h0_samples):
        test.append(np.random.normal(loc = 0, size = sample_size))
        labels.append(True)
    for _ in range(h1_samples):
        test.append(np.random.normal(loc = 0.5, size = sample_size))
        labels.append(False)
    return np.array(train), np.array(test), np.array(labels)
```

```python
def get_samples(h0_distribution=h0_distribution,
                h1_distribution=h1_distribution,
                sample_size = 1000,
                h0_samples = 500,
                h1_samples = 500):
    train = []
    test = []
    labels = []
    for _ in range(h0_samples):
        test.append(np.random.normal(loc = 0, size = sample_size))
        train.append(np.random.normal(loc = 0, size = sample_size))
        labels.append(True)
    for _ in range(h1_samples):
        test.append(np.random.normal(loc = 0.5, size = sample_size))
        train.append(np.random.normal(loc = 0, size = sample_size))
        labels.append(False)
    return np.array(train), np.array(test), np.array(labels)
```

```python
def apply_criterea(train, test, criterea = ttest_ind):
    try:
        assert(np.array(train).shape == np.array(test).shape)
        return np.array(map(lambda x: criterea(x[0], x[1])[1], zip(train, test)))
    except:
        return np.array(map(lambda x: criterea(train, x)[1], test))
```

```python
def simple_check(pvalues, alpha):
    return pvalues > alpha
```

```python
def check_results(pvalues, labels, alpha = 0.05, method = simple_check):
    m = len(pvalues)
    m0 = labels.sum()
    results = method(pvalues, alpha)
    U = (results * labels).sum()
    V = ((1 - results) *  labels).sum()
    T = (results * (1 - labels)).sum()
    S = ((1 - results) * (1 - labels)).sum()
    R = m - results.sum()
    table = pd.DataFrame(data = np.array([[U, T, m - R], [V, S, R], [m0, m - m0, m]]),
                         index = ["Принятых H0", "Отвергнутных Н0", "Всего"],
                         columns = ["Верных H0", "Неверных Н0", "Всего"])
    return table
```

```python
a,b,c = get_samples(sample_size=100)
apply_criterea(a, b)
check_results(apply_criterea(a, b), c)
```

```python
for _ in range(10):
    a,b,c = get_train_test_samples(sample_size=300)
    apply_criterea(a, b)
    print(check_results(apply_criterea(a, b), c))
```

```python
for _ in range(10):
    a,b,c = get_samples(sample_size=300)
    apply_criterea(a, b)
    print(check_results(apply_criterea(a, b), c))
```

### Метод Бонферрони

```python
def bonferoni(pvalues, alpha):
    return pvalues > alpha/len(pvalues)
for _ in range(10):
    a,b,c = get_samples(sample_size=300)
    apply_criterea(a, b)
    print(check_results(apply_criterea(a, b), c, method=bonferoni))
```

### Контролируем FWER - нисходящие процедуры

```python
def holm(pvalues, alpha):
    m = pvalues.shape[0]
    stop = m
    for i, pval in enumerate(sorted(pvalues)):
        if pval >= alpha/(m - i):
            stop = i
            break
    return np.argsort(pvalues)>=stop

for _ in range(10):
    a,b,c = get_train_test_samples(sample_size=300)
    apply_criterea(a, b)
    print(check_results(apply_criterea(a, b), c, method = holm))
```

```python
for _ in range(10):
    a,b,c = get_samples(sample_size=300)
    apply_criterea(a, b)
    print(check_results(apply_criterea(a, b), c, method = holm))
```

## Контролируем FDR - восходящие процедуры

```python
def Hochberg(pvalues, alpha):
    m = pvalues.shape[0]
    stop = -1
    for i, pval in enumerate(sorted(pvalues, reverse=True)):
        if pval < alpha*(i + 1)/m:
            stop = m - i
            break
    print(stop)
    return np.argsort(pvalues) >= stop

for _ in range(1):
    a,b,c = get_train_test_samples(sample_size=300)
    apply_criterea(a, b)
    print(check_results(apply_criterea(a, b), c, method = Hochberg))
```

```python
def Iekutelli(pvalues, alpha):
    m = pvalues.shape[0]
    stop = -1
    const = (1./(1+np.arange(m))).sum()
    for i, pval in enumerate(sorted(pvalues, reverse=True)):
        #print("pval {} comapred with {}".format(pval, alpha/((i + 1)*const)))
        if pval < alpha*(i+1)/(m*const):
            stop = m - i
            break
    print(stop)
    return np.argsort(pvalues) >= stop

for _ in range(100):
    a,b,c = get_samples(sample_size=300)
    apply_criterea(a, b)
    print(check_results(apply_criterea(a, b), c, method = Iekutelli))
```
