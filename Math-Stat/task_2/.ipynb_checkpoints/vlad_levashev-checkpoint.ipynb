```python
'''
Функция из нулевого домашнего задания.
''' 
def cumavg(temp):
    return np.divide(np.cumsum(temp, axis = 0), np.arange(1, len(temp)+1))
```

```python
evaluation_list = np.zeros((200,300))
evaluation = np.zeros((300))
for i in range(200):
    sample = sps.norm.rvs(size=300)
    evaluation_list[i] = cumavg(sample)
```

*a).* Параметрическая модель $\mathcal{N}(\theta, 1)$, выборки: [-1, 1], [-5, 5], [-1, 5]

*d).* Параметрическая модель $Bin(5, \theta)$, выборки: [0, 1], [5, 5], [0, 5]

*e).* Параметрическая модель $Pois(\theta)$, выборки: [0, 1], [0, 10], [5, 10]

*f).* Параметрическая модель $Сauchy(\theta)$, где $\theta$ --- параметр сдвига, выборки: [-0.5, 0.5], [-2, 2], [-4, 0, 4]

Выполнить задание, не создавая много кода, поможет следующая функция.

```python
def draw_likelihood(density_function, grid, samples, label):
    ''' density_function --- функция, считающая плотность (обычную или дискретную)
        grid --- сетка для построения графика
        samples --- три выборки
        label --- latex-код параметрической модели
    '''
    
    plt.figure(figsize=(18, 5))
    for i, sample in enumerate(samples):
        sample = np.array(sample)[np.newaxis, :]
        likelihood = np.prod(density_function(sample), axis=1)

        plt.subplot(1, 3, i+1)
        plt.plot(grid, likelihood)
        plt.xlabel('$\\theta$', fontsize=16)
        plt.grid(ls=':')
        plt.title(label + ', sample=' + str(sample), fontsize=16)
    plt.show()
```

Первый пункт можно выполнить с помощью следующего кода

```python
grid = np.linspace(-5, 5, 1000).reshape((-1, 1))
draw_likelihood(sps.norm(loc=grid).pdf, grid, 
                [[-1, 1], [-5, 5], [-1, 5]], '$\\mathcal{N}(\\theta, 1)$')
```

```python
grid = np.linspace(-5, 7, 1000).reshape((-1, 1))
draw_likelihood(sps.expon(scale=1/grid).pdf, grid,
               [[1, 2], [0.1, 1], [1, 10]], '$Exp(\\theta)$')
```

```python
grid = np.linspace(-5, 5, 1000).reshape((-1, 1))
draw_likelihood(sps.uniform(scale=grid).pdf, grid,
               [[0.2, 0.8], [0.5, 1], [0.5, 1.3]], '$U[0, \\theta]$')grid = np.linspace(-5, 5, 1000).reshape((-1, 1))
draw_likelihood(sps.binom(5, grid).pmf, grid,
               [[0, 1], [5, 5], [0, 5]], '$Bin(5, \\theta)$')
```

```python
grid = np.linspace(-5, 5, 1000).reshape((-1, 1))
draw_likelihood(sps.binom(5, grid).pmf, grid,
               [[0, 1], [5, 5], [0, 5]], '$Bin(5, \\theta)$')
```

```python
def cool_argmax(array):
    return np.unravel_index(np.argmax(array), array.shape)
```

```python
def likelihood(sample, c_, k_):
    return np.prod(sps.weibull_min.pdf(sample.reshape((sample.size,1,1)), c=c_, scale=k_), axis=0)
```

```python
precision = 0.1
x_left = 1.0
x_right = 15.0
y_left = 1.0
y_right = 20.0
grid = np.mgrid[x_left:x_right:precision, y_left:y_right:precision]
x = likelihood(data, grid[0], grid[1])
for i in range(4):
    y = cool_argmax(x)
    print(y)
    temp_x_left = x_left
    temp_y_left = y_left
    x_left = max(x_left, x_left + precision*(y[0]-2))
    x_right = min(x_right, temp_x_left + precision*(y[0]+2))
    y_left = max(y_left, y_left + precision*(y[1]-2))
    y_right = min(y_right, temp_y_left + precision*(y[1]+2))
    
    precision /= 10
    grid = np.mgrid[x_left:x_right:precision, y_left:y_right:precision]
    x = likelihood(data, grid[0], grid[1])
y = cool_argmax(x)
print(x_left, x_right)
print(y_left, y_right)
print(precision)
print(y)
print(x_left + y[0]*precision, y_left + y[1]*precision)
```
