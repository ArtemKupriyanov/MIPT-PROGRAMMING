**Задача 1.** В этой задаче нужно визуализировать свойство несмещенности.

Пусть $X_1, ..., X_n$ --- выборка из распределения $U[0, \theta]$. Известно, что в качестве оценки параметра $\theta$ можно использовать следующие оценки $X_{(n)}, \frac{n+1}{n}X_{(n)}, 2\overline{X}$.

**Вопрос:** Какие из этих оценок являются несмещенными?

**Ответ:** $\mathbb{E}X_{(n)} = \frac{n}{n + 1} \theta$; $\mathbb{E}\frac{n + 1}{n} X_{(n)} = \theta$; $\mathbb{E}2 \overline{X} = \theta$. Стало быть, первая оценка является смещённой (но асимптотически несмещённой, так как $\displaystyle{\lim_{n \to \infty} \frac{n}{n + 1} = 1}$, а последние две оценки -- несмещённые.

Вам нужно убедиться в этом, сгенерировав множество выборок и посчитав по каждой из них оценку параметра.

Сгенерируйте 500 выборок $X_1^j, ..., X_n^j$ из распределения $U[0, 1]$, по каждой из них посчитайте оценку $\widehat{\theta}_j$, получив тем самым 500 независимых оценок параметра. Нанесите их на график с одинаковой *у*-координатой. Отметьте специальным символом среднее этих выборок (см. шаблон ниже). Выполните данную процедуру для $n \in \{10, 100, 500\}$.

Для нанесения точек на график используйте следующий шаблон. Для каждой оценки выставите разный *уровень*, чтобы реализации разных оценок не слипались. В качестве *метки* используйте latex-код этой оценки, который можно взять выше в условии этой задачи. Постарайтесь не размножать код, а сделать циклы по типам оценок и по размеру выборки. Естественно, все типы оценок должны быть на одном графике, но для разных $n$ должны быть разные графики.

```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
import math
from statsmodels.distributions.empirical_distribution import ECDF
from statsmodels.nonparametric.kde import KDEUnivariate

%matplotlib inline
```

```python
ns = np.array([10, 100, 500])
size = 500

def eval1(X):
    return np.max(X, axis=1)

def eval2(X):
    n = X.shape[1]
    return np.max(X, axis=1) * (n + 1) / n

def eval3(X):
    return np.mean(X, axis=1) * 2

evals = np.array([eval1, eval2, eval3])
colors = np.array(["Blue", "Green", "Black"])
labels = np.array([r"$X_{(n)}$", r"$\frac{n + 1}{n} X_{(n)}$", r"$2 \overline{X}$"])
levels = np.array([10, 5, 0])

for n in ns:
    X = np.random.uniform(size=(n, size))
    
    plt.figure(figsize=(10, 3))
    plt.xlim((0.9, 1.1))
    
    for f, color, label, level in zip(evals, colors, labels, levels):
        evaluation = f(X)
        plt.scatter(evaluation, np.zeros_like(evaluation) + level,
                   alpha=0.4, s=100, color=color, label=label)
        plt.scatter(evaluation.mean(), level, marker='*', s=200,
                color='w', edgecolors='black')
        
    plt.vlines(1, -1, 11, color='r')
    plt.title('sample size = %d' % n)
    plt.yticks([])
    plt.legend()
    plt.show()   
```

Пусть теперь $X_1, ..., X_n$ --- выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$.

**Вопрос:** Какие из этих оценок являются несмещенными?

**Ответ:** $S^{2} = \overline{X^{2}} - (\overline{X})^{2}$. $\mathbb{E}S^{2} = \mathbb{E}\overline{X^{2}} - \mathbb{E}(\overline{X})^{2} = \frac{1}{n} \mathbb{E} (X_{1}^{2} + \ldots + X_{n}^2) - \frac{1}{n^{2}} \mathbb{E}(X_{1}^{2} + \ldots + X_{n}^{2}) - \frac{1}{n^{2}} \displaystyle{\sum_{i \neq j}} \mathbb{E} X_{i} X_{j} = \frac{n - 1}{n} \mathbb{E} X^{2} - \frac{n - 1}{n} (\mathbb{E} X)^{2} = \frac{n - 1}{n} \mathbb{D} X$. Стало быть, $S^{2}$ -- смещённая (но асимптотически несмещённая), а $\frac{n}{n - 1}S^{2}$ (исправленная выборочная дисперсия) -- несмещённая оценка параметра $\sigma^{2}$.

Для данной модели выполните те же действия, что и с предыдущей.

```python
ns = np.array([10, 100, 500])
size = 500

def eval1(X):
    return np.mean(X ** 2, axis=1) - np.mean(X, axis=1) ** 2

def eval2(X):
    n = X.shape[1]
    return eval1(X) * n / (n - 1)

evals = np.array([eval1, eval2])
colors = np.array(["Blue", "Green"])
labels = np.array([r"$S^{2}$", r"$\frac{n}{n - 1}S^{2}$"])
levels = np.array([5, 0])

# sigma2 = 25

for n in ns:
    X = np.random.normal(size=(n, size), scale=5)
    
    plt.figure(figsize=(10, 3))
    plt.xlim((20, 30))
    
    for f, color, label, level in zip(evals, colors, labels, levels):
        evaluation = f(X)
        plt.scatter(evaluation, np.zeros_like(evaluation) + level,
                   alpha=0.4, s=100, color=color, label=label)
        plt.scatter(evaluation.mean(), level, marker='*', s=200,
                color='w', edgecolors='black')
        
    plt.vlines(25, -1, 6, color='r')
    plt.title('sample size = %d' % n)
    plt.yticks([])
    plt.legend()
    plt.show()   
```

Сделайте вывод о том, что такое свойство несмещенности. Подтверждают ли сделанные эксперименты свойство несмещенности данных оценок? Поясните, почему в лабораторных по физике при оценке погрешности иногда используют $n-1$ в знаменателе, а не $n$.

**Вывод:** Несмещённость оценки говорит о том, что при большом количество замеров (выборок) среднее значение (математическое ожидание) оценки совпадает со значением оцениваемого параметра. Эксперименты подтверждают свойства несмещённости оценок при большом количестве выборок (при малых $n$ может с достаточно большой вероятностью произойти что угодно). В физике иногда используют знаменатель $n - 1$ при подсчёте погрешностей, так как именно такая формула даёт несмещённую оценку дисперсии. Однако зачастую используется знаменатель $n$ (и выборочная дисперсия), считая значение $\sqrt\frac{n}{n - 1}$ мало отличающимся от единицы.

По каждой из них посчитайте оценки $\widehat{\theta}_{jn} = \frac{1}{n}\sum\limits_{i=1}^n X_i^j$ для $1 \leqslant n \leqslant 300$, то есть оценка параметра по первым $n$ наблюдениям $j$-й выборки. **Задача 3.** В этой задаче нужно визуализировать свойство асимптотической нормальности.

*a).* Пусть $X_1, ..., X_n$ --- выборка из распределения $\mathcal{N}(\theta, 1)$. Известно, что $\overline{X}$ является асимптотически нормальной оценкой параметра $\theta$. Вам нужно убедиться в этом, сгенерировав множество выборок и посчитав по каждой из них оценку параметра в зависимости от размера выборки.

Сгенерируйте 200 выборок $X_1^j, ..., X_{300}^j$ из распределения $\mathcal{N}(0, 1)$. По каждой из них посчитайте оценки $\widehat{\theta}_{jn} = \frac{1}{n}\sum\limits_{i=1}^n X_i^j$ для $1 \leqslant n \leqslant 300$, то есть оценка параметра по первым $n$ наблюдениям $j$-й выборки. Для этой оценки посчитайте статистику $T_{jn} = \sqrt{n} \left( \widehat{\theta}_{jn} - \theta \right)$, где $\theta = 0$.

Для каждого $j$ нанесите на один график зависимость $T_{jn}$ от $n$ с помощью `plt.plot`. Каждая кривая должна быть нарисована *одним цветом* с прозрачностью `alpha=0.2`. Сходятся ли значения $T_{jn}$ к какой-либо константе?

Для $n=300$ по выборке $T_{1,300}, ..., T_{200,300}$ постройте гистограмму и ядерную оценку плотности. Хорошо ли они приближают плотность распределения $\mathcal{N}(0, 1)$ (ее тоже постройте на том же графике)? Не забудьте сделать легенду.

```python
size = 300

sample = np.zeros(200)

plt.figure(figsize=(10, 3))

for n in range(200):
    X = np.random.normal(size=size)
    evaluation = X.cumsum() / np.arange(1, size + 1, 1)
    T = evaluation * np.sqrt(np.arange(1, size + 1, 1))
    plt.plot(T, color="Blue", alpha=0.2)
    sample[n] = T[size - 1]
    
plt.ylim((-5, 5))
plt.show()

plt.figure(figsize=(10, 3))

# sample = np.random.normal(size=200)

grid = np.linspace(-3, 3, 200)
plt.hist(sample, bins=30, range=(grid.min(), grid.max()), normed=True)

kernel_density = KDEUnivariate(sample)
kernel_density.fit()
        
plt.plot(grid, kernel_density.evaluate(grid),
        color='Green', label='kde')
plt.plot(grid, sps.norm.pdf(grid),
        color='Red', label='true pdf')

plt.legend()
plt.show()
```

$T_{jn}$ не очень-то куда-то сходится, но это и правильно. По ЦПТ, $\sqrt{n} (\theta^{*}_{n} - \theta) \overset{\text{d}}{\to} \xi \sim \mathcal{N}(0, 1)$. То есть $T_{jn}$ должно всё более и более походить на стандартное нормальное распределение, что и происходит. Ядерная оценка плотности выборки $T_{j, 300}$ очень похожа на плотность стандартного нормального распределения, что и ожидалось.

*b).* Пусть $X_1, ..., X_n$ --- выборка из распределения $Pois(\theta)$. Известно, что $\overline{X}$ является асимптотически нормальной оценкой параметра $\theta$. Выполните исследование, аналогичное пункту *a)*.

```python
size = 300

lam = 3

sample = np.zeros(200)
plt.figure(figsize=(10, 3))

for n in range(200):
    X = np.random.poisson(size=size, lam=lam)
    evaluation = X.cumsum() / np.arange(1, size + 1, 1)
    T = (evaluation - lam) * np.sqrt(np.arange(1, size + 1, 1))
    plt.plot(T, color="Blue", alpha=0.2)
    sample[n] = T[size - 1]
    
plt.ylim((-7, 7))
plt.show()

plt.figure(figsize=(10, 3))

grid = np.linspace(-7, 7, 200)
plt.hist(sample, bins=30, range=(grid.min(), grid.max()), normed=True)

kernel_density = KDEUnivariate(sample)
kernel_density.fit()
        
plt.plot(grid, kernel_density.evaluate(grid),
        color='Green', label='kde')
plt.plot(grid, sps.norm(scale=math.sqrt(lam)).pdf(grid),
        color='Red', label='true pdf')

plt.legend()
plt.show()
```

Вновь по ЦПТ $\sqrt{n}(\theta^{*}_{n} - \theta) \overset{\text{d}}{\to} \mathcal{N}(0, \sqrt{\theta})$, в данном случае $\theta = 3$. Ядерная оценка плотности похожа на плотность стандартного с нулевым сдвигом и $\sigma^{2} = 3$.

Сделайте вывод о том, что такое свойство асимптотической нормальности. Подтверждают ли сделанные эксперименты свойство асимптотической нормальности данных оценок? Как связаны результаты с центральной предельной теоремой?

Асимптотическая нормальность оценки $\theta^{*}_{n}$ говорит о том, что случайная величина $\sqrt{n}(\theta^{*}_{n} - \theta)$ стремится по распределению к случайной величине с нормальным распределением с нулевым смещением. То есть при больших $n$ выборка из $\sqrt{n}(\theta^{*}_{n} - \theta)$ похожа на выборку из нормального распределения. Эксперименты подтверждают свойство асимптотической нормальности приведённых оценок. Для доказательства асимптотической нормальности можно пользоваться ЦПТ.

**Задача 4.** Пусть $X_1, ..., X_n$ --- выборка из распределения $U[0, \theta]$. Из домашнего задания известно, что $n\left(\theta - X_{(n)}\right) \stackrel{d_\theta}{\longrightarrow} Exp\left(1/\theta\right)$. Проведите исследование, аналогичное заданию 3 для $\theta=1$.

```python
size = 300

plt.figure(figsize=(10, 6))
sample = np.zeros(200)

for n in range(200):
    X = np.random.uniform(size=size)
    evaluation = np.maximum.accumulate(X)
    T = np.arange(1, size + 1, 1) * (1 - evaluation)
    plt.plot(T, color="Blue", alpha=0.2)
    sample[n] = T[size - 1]
    
plt.ylim((-1, 8))
plt.show()

grid = np.linspace(0, 4, 200)
plt.figure(figsize=(10, 3))
plt.hist(sample, bins=30, range=(grid.min(), grid.max()), normed=True)

kernel_density = KDEUnivariate(sample)
kernel_density.fit()
        
plt.plot(grid, kernel_density.evaluate(grid),
        color='Green', label='kde')
plt.plot(grid, sps.expon.pdf(grid),
        color='Red', label='true pdf')

plt.legend()
plt.show()
```

Практика в очередной раз сошлась с теорией.

**Задача 5.** Дана параметрическая модель и несколько выборок из двух или трех наблюдений (для удобства они даются в виде python-кода). Нужно для каждой выборки построить график функции правдоподобия.

*a).* Параметрическая модель $\mathcal{N}(\theta, 1)$, выборки: [-1, 1], [-5, 5], [-1, 5]

*b).* Параметрическая модель $Exp(\theta)$, выборки: [1, 2], [0.1, 1], [1, 10]

*c).* Параметрическая модель $U[0, \theta]$, выборки: [0.2, 0.8], [0.5, 1], [0.5, 1.3]

*d).* Параметрическая модель $Bin(5, \theta)$, выборки: [0, 1], [5, 5], [0, 5]

*e).* Параметрическая модель $Pois(\theta)$, выборки: [0, 1], [0, 10], [5, 10]

*f).* Параметрическая модель $Сauchy(\theta)$, где $\theta$ --- параметр сдвига, выборки: [-0.5, 0.5], [-2, 2], [-4, 0, 4]

Выполнить задание, не создавая много кода, поможет следующая функция.

```python
def draw_likelihood(density_function, grid, samples, label):
    ''' density_function --- функция, считающая плотность (обычную или дискретную)
        grid --- сетка для построения графика
        samples --- три выборки
        label --- latex-код параметрической модели
    '''
    
    plt.figure(figsize=(18, 5))
    for i, sample in enumerate(samples):
        sample = np.array(sample)[np.newaxis, :]
        likelihood = density_function(sample).prod(axis=1)

        plt.subplot(1, 3, i+1)
        plt.plot(grid, likelihood)
        plt.xlabel('$\\theta$', fontsize=16)
        plt.grid(ls=':')
        plt.title(label + ', sample=' + str(sample), fontsize=16)
    plt.show()
```

Сделайте вывод о том, как функция правдоподобия для каждой модели зависит от выборки. Является ли функция правдоподобия плотностью?

Функция правдоподобия есть функция от $\theta$ при фиксированной выборке. Функция правдоподобия в точке $\theta$ есть произведение значений плотности распределения $P_{\theta}$ в точках выборки. Функция правдоподобия не является плотностью, так как является произведением значений плотностей. Да, некоторые графики выглядят похоже на плотность соответствующего распределения, но это вызвано свойствами плотности распределений.

 Первый пункт можно выполнить с помощью следующего кода.

```python
# Пункт a

grid = np.linspace(-5, 5, 1000).reshape((-1, 1))
draw_likelihood(sps.norm(loc=grid).pdf, grid, 
                [[-1, 1], [-5, 5], [-1, 5]], '$\\mathcal{N}(\\theta, 1)$')
```

Функция правдоподобия в этой модели имеет вид $~f_{\theta} = De^{A\theta^{2} + B\theta + C}$, что очень похоже не плотность нормального распределения, но коэффициент $D$ не позволяет выбрать коэффициенты смещения и масштаба, чтобы это действительно было плотностью. (Здесь и далее подразумевается, что A, B, C, и так далее, есть функции от X, но нас интересует лишь их значение при определённой выборке).

```python
# Пункт b

grid = np.linspace(-1, 10, 1000).reshape((-1, 1))

draw_likelihood(sps.expon(scale=1 / grid).pdf, grid,
                [[1, 2], [0.1, 1], [1, 10]], r"Exp($\theta$)")
```

$f_{\theta}(X) = \theta^{n} e^{-\theta(X_{1} + \ldots + X_{n})}$. Это не плотность, так как коэффициент перед экспонентой должен быть постоянным, но он зависит от аргумента. Однако формой функция распределения похожа на плотность экспоненциального распределения, что и видно на графике.

```python
# Пункт c

grid = np.linspace(-1, 10, 1000).reshape((-1, 1))

draw_likelihood(sps.uniform(scale=grid).pdf, grid,
                [[0.2, 0.8], [0.5, 1], [0.5, 1.3]], r"U[0, $\theta$")
```

$f_{\theta} = \frac{1}{\theta^{n}}I(\theta \geq X_{(n)})$. В отличие от плотности равномерного распределения $p_{\theta}(x) = \frac{1}{\theta}I(0 \leq x \leq \theta)$ в формуле для функции правдоподобия нет индикатора, ограничивающего $\theta$ сверху, поэтому её носитель (множество $\theta$, в которых функция правдоподобия не ноль) бесконечен.

```python
# Пункт d

grid = np.linspace(-1, 2, 1000).reshape((-1, 1))

draw_likelihood(sps.binom(n=5, p=grid).pmf, grid,
                [[0, 1], [5, 5], [0, 5]], r"Bin(5, $\theta$)")
```

Функция правдоподобия есть $C \theta^{s} (1 - \theta)^{5n - s}$, где $n$ -- размер выборки.

```python
# Пункт e

grid = np.linspace(-1, 15, 1000).reshape((-1, 1))

draw_likelihood(sps.poisson(grid).pmf, grid,
                [[0, 1], [0, 10], [5, 10]], r"Pois($\theta$)")
```

$f_{\theta}(X) = C e^{-n\theta} \theta^{X_{1} + \ldots + X_{n}}$, что похоже на плотность.

```python
# Пункт f

grid = np.linspace(-5, 5, 1000).reshape((-1, 1))

draw_likelihood(sps.cauchy(loc=grid).pdf, grid,
                [[-0.5, 0.5], [-2, 2], [-4, 0, 4]], r"Cauchy($\theta$)")
```

Последние два графика совершенно не похожи на график плотности распределения Коши.

**Вывод:** Функция правдоподобия показывает, насколько хороши все возможные оценки параметра $\theta$ при фиксированной выборке (чем больше значение в точке $\theta$, тем более правдоподобен именно такой исход). В общем случае она не связана с плотностью соответствующего распределения.

Сгенерируем выборку большого размера из стандартного нормального распределения и посчитаем ее функцию правдоподобия в модели $\mathcal{N}(\theta, 1)$. Выполните код ниже

```python
sample = sps.norm.rvs(size=10**5)

for loc in range(-5, 5):
    likelihood = sps.norm(loc=loc).pdf(sample).prod()
    print(loc, likelihood)
```

Почему результат отличается от ожидаемого? Как обойти эту неприятность для подсчтета оценки максимального правдоподобия? Реализуйте это. *Подсказка:* нужно использовать некоторую функцию у класса, который реализует это распределения.

Функция правдоподобия в точке $\theta$ при фиксированной выборке есть величина, мало отличающаяся от нуля, поэтому мы не можем понять, какая оценка лучше: $\theta^{*} = 0$ или $\theta^{*} = 5$. Используем логарифмическую функцию правдоподобия.

```python
for loc in range(-5, 5):
    likelihood = np.log(sps.norm(loc=loc).pdf(sample)).sum()
    print(loc, likelihood)
```

Получили, что значение функции правдоподобия в точке $\theta = 0$ похоже на максимум (среди построенных), поэтому настоящая ОМП для параметра $\theta$ в данной модели есть $\theta^{*} \approx 0$.
